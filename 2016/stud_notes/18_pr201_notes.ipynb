Вспомогательная задача от Бориса Борисовича( к 15.5 с листочка)
-------------

Из Москвы в Самару летят 2 самолета. В 1-ом заняты все места, во 2-ом только одно. В самолёте 200 мест.
Мы ловим в аэропорту случайно одного из двух пилотов и спрашиваем его, сколько пассажиров летело на его самолёте. Посчитаем математическое ожидание этой случайной величины: 
    
$$
E(X)=0.5 \times 1+0.5 \times 200 = 100.5
$$



Перекинувшись с пилотом еще парочкой фраз, мы отправляемся гулять по аэропорту, 
где удивительным образом находятся пассажиры исключительно с рейса "Москва-Самара".
Из любопытства мы задаем им тот же вопрос, что и пилоту. Математическое ожидание этой новой случайной величины рассчитывается так: 

$$
E(Y)=\frac{200}{201} \times 200+ \frac{1}{201} \times 1 = 199 
$$

**Решение от Полины.**

Имеются три неравенства:


$ \frac{Y_{BM}}{M_{B}}>\frac{Y_{АМ}}{M_{A}} $
                                   


$\frac{Y_{ВЖ}}{Ж_{B}}>\frac{Y_{АЖ}}{Ж_{A}}$

$\frac{Y_{АЖ}+Y_{AM}}{А}>\frac{Y_{BЖ}+Y_{BM}}{B}$

где М-мужчины, принимавшие лекарство (индекс указывает на лекарство: А или В)

Ж-женщины, принимавшие лекарство (аналогично)

У-улучшение (первая буква индекса указывает на лекарство, вторая- на группу (мужчины/женщины), которая почувствовала улучшение после его приема)

Пояснение:
$M_{A}+Ж_{A}=A$,
$M_{B}+Ж_{B}=B$

*Применяем немного магии, много подбора и приходим к бобрически-бомбическому решению.*

**Метод Моментов**: по ЗБЧ, эмпирическое среднее (среднее арифметическое) достаточно большой конечной выборки из фиксированного распределения близко к теоретическому среднему (математическому ожиданию) этого распределения.

100-491$\hat{a}$=$\bar{y}$

*Estimator:*
$\hat{a}=\frac{100- \bar{y}}{491}$

$\bar{y}=\frac{2+2+100+1}{4}=26.25$, следовательно, $\hat{a}=\frac{100-26.25}{491} \approx 0.15$
